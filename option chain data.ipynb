```python
import requests
import pandas as pd
import xlwings as xw
import time
```

```python
url = 'https://www.nseindia.com/api/option-chain-indices?symbol=NIFTY'
url2 = 'https://www.nseindia.com/api/option-chain-indices?symbol=BANKNIFTY'
headers = {'accept-encoding': 'gzip, deflate, br',
            'accept-language': 'en-GB,en-US;q=0.9,en;q=0.8',
           'user-agent': 'Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'
          }
```

```python
while True:
    try:
        session = requests.session()
        dataNifty = session.get(url,headers=headers).json()['records']['data']
        dataBankNifty = session.get(url2,headers=headers).json()['records']['data']
        ocdataNifty = []
        ocdataBankNifty = []
        for i in dataNifty:
            for j,k in i.items():
                if j=='CE' or j=='PE':
                    infoNifty = k
                    infoNifty['instrumentType']=j
                    ocdataNifty.append(infoNifty)
        dfNifty = pd.DataFrame(ocdataNifty, columns=['strikePrice', 'expiryDate','openInterest','changeinOpenInterest','pchangeinOpenInterest','totalTradedVolume','impliedVolatility','lastPrice','change','pChange',
                                          'totalBuyQuantity','totalSellQuantity','bidQty','bidprice','askQty','askPrice','underlyingValue',
                                          'instrumentType'])
        for i in dataBankNifty:
            for j,k in i.items():
                if j=='CE' or j=='PE':
                    infoBankNifty = k
                    infoBankNifty['instrumentType']=j
                    ocdataBankNifty.append(infoBankNifty)
        dfBankNifty = pd.DataFrame(ocdataBankNifty, columns=['strikePrice', 'expiryDate','openInterest','changeinOpenInterest','pchangeinOpenInterest','totalTradedVolume','impliedVolatility','lastPrice','change','pChange',
                                          'totalBuyQuantity','totalSellQuantity','bidQty','bidprice','askQty','askPrice','underlyingValue',
                                          'instrumentType'])

        ExpryNifty = dfNifty[dfNifty['expiryDate'] == '29-Dec-2022']
        ceNifty = ExpryNifty[ExpryNifty['instrumentType'] == 'CE']
        peNifty =  ExpryNifty[ExpryNifty['instrumentType'] == 'PE']

        ExpryBankNifty = dfBankNifty[dfBankNifty['expiryDate'] == '22-Dec-2022']
        ceBankNifty = ExpryBankNifty[ExpryBankNifty['instrumentType'] == 'CE']
        peBankNifty =  ExpryBankNifty[ExpryBankNifty['instrumentType'] == 'PE']

        wb = xw.Book('E:\Project practice\option Chain.xlsm')
        dfNifty = wb.sheets('RawDataNifty')
        dfNifty.range('A2').value = ceNifty
        dfNifty.range('V2').value = peNifty
        dfBankNifty = wb.sheets('RawDataBankNifty')
        dfBankNifty.range('A2').value = ceBankNifty
        dfBankNifty.range('V2').value = peBankNifty
        time.sleep(20)
    except Exception as e:
        print(e)
```
